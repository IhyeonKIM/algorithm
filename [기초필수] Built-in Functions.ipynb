filter

```python
def myfun(val):
  if val %2 ==0:
    return True
  else: return None 

list(filter(myfun, range(20))) # Filter 이용해서 어떤 값들을 특정 함수로 적용시켜주는 방법 
```

```python
list(filter(lambda x : x %2 ==0 , range(20)  ) ) 
```

```python
[i for i in range(20) if i % 2 ==0 ]
```

```python
testCaseOne = [ 'abc' , 'de' ,'helloworld','hello','python']
testCaseTwo = "Life is too short, You need python".split()
testCaseThree = list(zip(['w','c','a','j'] , [1,2,3,4],[8,7,6,5]))

sorted(testCaseOne, key = len , reverse = True) # reverse 옵션때문에 긴 것 먼저 나옴. 
```

```python
sorted(testCaseThree, key = lambda x:x[1], reverse = True ) # 인덱스 1의 리버스. 
sorted(testCaseThree, key = lambda x:x[0] ) # 인덱스 1의 알파벳순 
```

- 단톡방에 x마리 동물이 대화를 하고 있다.
- 각각의 동물들이 톡을 전송할 때마다 서버에는 아래와 같이 저장된다.
- 1. 단톡방에는 모두 몇마리의 동물이 있을까요? 톡은 무조건 1회이상 전송합니다
- 2. 단톡방에 동물들마다 몇번의 톡을 올렸을까요?

```python
serverData= '개리 라이캣 개리 개리 라이캣 자바독 자바독 파이 썬'
```

```python
data = set( serverData.split( ) ) 
print(data)
print( '총' , len(data), ' 마리 ')
```

```python
d = {}
for i in data:
  print( i, serverData.split( ).count(i))
  d[i] = serverData.split().count(i)

d
```
